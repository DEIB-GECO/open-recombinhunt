```python
import pandas as pd
```

```python
df = pd.read_csv("../../results/nextstrain_output/sars-cov-2/nextstrain_reformatted.tsv", sep='\t')
```

```python
output_file_last_6_months = "../../results/nextstrain_output/sars-cov-2/nextstrain_reformatted_last_6_months.tsv"
```

```python
current_date = pd.Timestamp.now()
six_months_ago = current_date - pd.DateOffset(months=7)
df_last_6_months = df[df['Collection date'] >= six_months_ago.strftime('%Y-%m-%d')]
df_last_6_months.to_csv(output_file_last_6_months, sep='\t', index=False)
```
