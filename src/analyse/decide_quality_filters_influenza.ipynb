IT IS BEST TO HANDLE SEGMENT COLUMN BY A SCRIPT -AS THIS COLUMN IS NOT GOOD QUALITY.
KEEP:

- 4.0
- 4
- HA
- segment 4

```python
# filter Segment = [4.0, 4, HA, segment 4] and Geographic Region = [North America]
df_filtered1 = df[(df["Geographic Region"] == "North America")]
df_filtered2 = df_filtered1[(df_filtered1["Segment"].isin(["4.0", "4", "HA", "segment 4"]))]

len(df_filtered2)
```

```python
# organism name contains H5N1 or h5n1
df_filtered3 = df[df["Virus Name"].str.contains("H5N1|h5n1", na=False)]

print(f"Number of records after filtering for H5N1: {len(df_filtered3)}")

df_filtered4 = df_filtered3[df_filtered3["Segment"].isin(["4.0", "4", "HA", "segment 4"])]
df_filtered5 = df_filtered4[df_filtered4["Geographic Region"] == "North America"]

print(f"Number of records after filtering: {len(df_filtered5)}")
```
