```python
virus = "rsv-b"
dist = 5
size = 50
```

```python
import pandas as pd
import os
```

```python
path = f"../../environments/{virus}/dist{dist}size{size}/"
files = os.listdir(path)
```

```python
for file in files:
    # read parquet file
    df = pd.read_parquet(os.path.join(path, file))
    print("environment file name:", file, "\nshape:", df.shape)
    print()
    print(df.head(50))
    print("\n\n")
```

```python
# print every element in the dataframe one by one
for index, row in df.iterrows():
    print(f"{row.to_dict()}")
```
