```python
viruses = [
    "yellow-fever",
    "zika",
    "influenza",
    "monkeypox",
    "rsv-a",
    "rsv-b",
]
```

```python
# load reference.fasta
# ../../data/processed/{virus}/reference.fasta
# get length of reference sequence
# using BioPython
from Bio import SeqIO
def get_length(fasta):
    for record in SeqIO.parse(fasta, "fasta"):
        return len(record.seq)
```

```python
for v in viruses:
    path = f"../../data/processed/{v}/reference.fasta"
    print(f"{v}: length of the reference sequence:", get_length(path))
```
